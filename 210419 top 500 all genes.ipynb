```python
IgE_over_unstim = pd.Index(df['M IgE/unstim LFC>']).intersection(pd.Index(df['F IgE/unstim LFC>'])).to_frame(index=False, name='IgE>unstim').dropna()
unstim_over_IgE = pd.Index(df['M IgE/unstim LFC<']).intersection(pd.Index(df['F IgE/unstim LFC<'])).to_frame(index=False, name='IgE<unstim').dropna()
M_over_F = pd.Index(df['M/F unstim LFC>']).intersection(pd.Index(df['M/F IgE LFC>'])).to_frame(index=False, name='M>F').dropna()
F_over_M = pd.Index(df['M/F unstim LFC<']).intersection(pd.Index(df['M/F IgE LFC<'])).to_frame(index=False, name='M<F').dropna()
intersect = pd.concat([M_over_F, F_over_M, IgE_over_unstim, unstim_over_IgE], axis=1)
```

```python
unstim_M_over_F = pd.Index(df['M/F unstim LFC>']).difference(pd.Index(df['M/F IgE LFC>'])).to_frame(index=False, name='unstim M>F').dropna()
unstim_F_over_M = pd.Index(df['M/F unstim LFC<']).difference(pd.Index(df['M/F IgE LFC<'])).to_frame(index=False, name='unstim M<F').dropna()
ige_M_over_F = pd.Index(df['M/F IgE LFC>']).difference(pd.Index(df['M/F unstim LFC>'])).to_frame(index=False, name='IgE M>F').dropna()
ige_F_over_M = pd.Index(df['M/F IgE LFC<']).difference(pd.Index(df['M/F unstim LFC<'])).to_frame(index=False, name='IgE M<F').dropna()
M_unstim_over_IgE = pd.Index(df['M IgE/unstim LFC>']).difference(pd.Index(df['F IgE/unstim LFC>'])).to_frame(index=False, name='M IgE>unstim').dropna()
M_IgE_over_unstim = pd.Index(df['M IgE/unstim LFC<']).difference(pd.Index(df['F IgE/unstim LFC<'])).to_frame(index=False, name='M IgE<unstim').dropna()
F_unstim_over_IgE = pd.Index(df['F IgE/unstim LFC>']).difference(pd.Index(df['M IgE/unstim LFC>'])).to_frame(index=False, name='F IgE>unstim').dropna()
F_IgE_over_unstim = pd.Index(df['F IgE/unstim LFC<']).difference(pd.Index(df['M IgE/unstim LFC<'])).to_frame(index=False, name='F IgE<unstim').dropna()
complements = pd.concat([unstim_M_over_F, unstim_F_over_M, ige_M_over_F, ige_F_over_M, M_unstim_over_IgE, M_IgE_over_unstim, F_unstim_over_IgE, F_IgE_over_unstim], axis=1)
```

```python
overlaps= pd.concat([intersect, complements], axis=1)
```
