```python
df1 = pd.read_excel("F_IgE_vs_M_IgE_Differential_Expression.xlsx")[['Entrez ID', 'F_IgE', 'M_IgE']]
df2 = pd.read_excel("F_Unstim_vs_F_IgE_Differential_Expression.xlsx")[['Entrez ID', 'F_Unstim', 'F_IgE']]
df3 = pd.read_excel("F_Unstim_vs_M_Unstim_Differential_Expression.xlsx")[['Entrez ID', 'F_Unstim', 'M_Unstim']]
df4 = pd.read_excel("M_Unstim_vs_M_IgE_Differential_Expression.xlsx")[['Entrez ID', 'M_Unstim', 'M_IgE']]
df5 = pd.read_excel("Unstim_vs_IgE_Differential_Expression.xlsx")[['Entrez ID', 'F_unstim_S4', 'M_unstim_S2', 'F_IgE_S3', 'M_IgE_S1']]
```

```python
rawmerged = pd.merge(left=df5, right=df4, left_on='Entrez ID', right_on='Entrez ID')
rawmerged = pd.merge(left=rawmerged, right=df3, left_on='Entrez ID', right_on='Entrez ID')
rawmerged = pd.merge(left=rawmerged, right=df2, left_on='Entrez ID', right_on='Entrez ID')
rawmerged = pd.merge(left=rawmerged, right=df1, left_on='Entrez ID', right_on='Entrez ID')
```

```python
rawmerged['M_Unstim_ave']=rawmerged[['M_Unstim_x', 'M_Unstim_y', 'M_unstim_S2']].mean(axis=1)
rawmerged['F_Unstim_ave']=rawmerged[['F_Unstim_x', 'F_Unstim_y', 'F_unstim_S4']].mean(axis=1)
rawmerged['F_IgE_ave']=rawmerged[['F_IgE_x', 'F_IgE_y', 'F_IgE_S3']].mean(axis=1)
rawmerged['M_IgE_ave']=rawmerged[['M_IgE_x', 'M_IgE_y', 'M_IgE_S1']].mean(axis=1)
```

```python
clean = rawmerged[['Entrez ID', 'F_Unstim_ave', 'M_Unstim_ave', 'F_IgE_ave', 'M_IgE_ave']]
clean.columns = ['Entrez', 'F_unstim', 'M_unstim', 'F_IgE', 'M_IgE']
clean =clean.mask(clean.eq(0)).dropna(thresh=2).fillna(1)
```

```python
eae=pd.read_excel('gene_result.xlsx')[['Org_name', 'GeneID', 'Symbol']]
eae=eae[eae['Org_name'].str.contains('Mus musculus')]
```

```python
df1=clean[clean['Entrez'].isin(eae['GeneID'].values)].merge(eae, left_on='Entrez', right_on='GeneID').drop(['Org_name', 'GeneID'], axis=1)
```

```python
df1=df1[['Entrez', 'Symbol', 'F_unstim', 'M_unstim', 'F_IgE', 'M_IgE']]
```

```python
df1
```

```python
df1['log2_F_unstim'] = np.log2(df1['F_unstim'])
df1['log2_M_unstim'] = np.log2(df1['M_unstim'])
df1['log2_F_IgE'] = np.log2(df1['F_IgE'])
df1['log2_M_IgE'] = np.log2(df1['M_IgE'])
```

```python
df1['L2FC M/F unstim'] = df1['log2_M_unstim'] - df1['log2_F_unstim']
df1['L2FC M/F IgE'] = df1['log2_M_IgE'] - df1['log2_F_IgE']
df1['L2FC M IgE/unstim'] = df1['log2_M_IgE'] - df1['log2_M_unstim']
df1['L2FC F IgE/unstim'] = df1['log2_F_IgE'] - df1['log2_F_unstim']
```

```python
df1.to_excel('210107 RNAseq eae-related genes.xlsx', index=False)
```

```python
unstimMoverF = df1.nlargest(40, 'L2FC M/F unstim', keep='all')['Symbol'].reset_index(drop=True)
unstimFoverM = df1.nsmallest(40, 'L2FC M/F unstim', keep='all')['Symbol'].reset_index(drop=True)
igeMoverF = df1.nlargest(40, 'L2FC M/F IgE', keep='all')['Symbol'].reset_index(drop=True)
igeFoverM = df1.nsmallest(40, 'L2FC M/F IgE', keep='all')['Symbol'].reset_index(drop=True)
MIgEoverunstim = df1.nlargest(40, 'L2FC M IgE/unstim', keep='all')['Symbol'].reset_index(drop=True)
MunstimoverIgE = df1.nsmallest(40, 'L2FC M IgE/unstim', keep='all')['Symbol'].reset_index(drop=True)
FIgEoverunstim = df1.nlargest(40, 'L2FC F IgE/unstim', keep='all')['Symbol'].reset_index(drop=True)
FunstimoverIgE = df1.nsmallest(40, 'L2FC F IgE/unstim', keep='all')['Symbol'].reset_index(drop=True)
```

```python
top40 = pd.DataFrame({'unstim_M_over_F':unstimMoverF, 'unstim_F_over_M':unstimFoverM, 'ige_M_over_F':igeMoverF, 'ige_F_over_M':igeFoverM, 'M_unstim_over_IgE':MunstimoverIgE, 'M_IgE_over_unstim':MIgEoverunstim, 'F_unstim_over_IgE':FunstimoverIgE, 'F_IgE_over_unstim':FIgEoverunstim})
```

```python
top40
```

```python
M_over_F = pd.Index(top40['unstim_M_over_F']).intersection(pd.Index(top40['ige_M_over_F'])).to_frame(index=False, name='M_over_F').dropna()
F_over_M = pd.Index(top40['unstim_F_over_M']).intersection(pd.Index(top40['ige_F_over_M'])).to_frame(index=False, name='F_over_M').dropna()
Ige_over_unstim = pd.Index(top40['M_IgE_over_unstim']).intersection(pd.Index(top40['F_IgE_over_unstim'])).to_frame(index=False, name='Ige_over_unstim').dropna()
unstim_over_IgE = pd.Index(top40['M_unstim_over_IgE']).intersection(pd.Index(top40['F_unstim_over_IgE'])).to_frame(index=False, name='unstim_over_IgE').dropna()
intersects = pd.concat([M_over_F, F_over_M, Ige_over_unstim, unstim_over_IgE], axis=1)
```

```python
unstim_M_over_Fa = pd.Index(top40['unstim_M_over_F']).difference(pd.Index(top40['ige_M_over_F'])).to_frame(index=False, name='unstim_M_over_Fa').dropna()
unstim_F_over_Ma = pd.Index(top40['ige_M_over_F']).difference(pd.Index(top40['unstim_M_over_F'])).to_frame(index=False, name='unstim_F_over_Ma').dropna()
ige_M_over_Fa = pd.Index(top40['unstim_F_over_M']).difference(pd.Index(top40['ige_F_over_M'])).to_frame(index=False, name='ige_M_over_Fa').dropna()
ige_F_over_Ma = pd.Index(top40['ige_F_over_M']).difference(pd.Index(top40['unstim_F_over_M'])).to_frame(index=False, name='ige_F_over_Ma').dropna()
M_unstim_over_IgEa = pd.Index(top40['M_IgE_over_unstim']).difference(pd.Index(top40['F_IgE_over_unstim'])).to_frame(index=False, name='M_unstim_over_IgEa').dropna()
M_IgE_over_unstima = pd.Index(top40['F_IgE_over_unstim']).difference(pd.Index(top40['M_IgE_over_unstim'])).to_frame(index=False, name='M_IgE_over_unstima').dropna()
F_unstim_over_IgEa = pd.Index(top40['M_unstim_over_IgE']).difference(pd.Index(top40['F_unstim_over_IgE'])).to_frame(index=False, name='F_unstim_over_IgEa').dropna()
F_IgE_over_unstima = pd.Index(top40['F_unstim_over_IgE']).difference(pd.Index(top40['M_unstim_over_IgE'])).to_frame(index=False, name='F_IgE_over_unstima').dropna()
complements = pd.concat([unstim_M_over_Fa, unstim_F_over_Ma, ige_M_over_Fa, ige_F_over_Ma, M_unstim_over_IgEa, M_IgE_over_unstima, F_unstim_over_IgEa, F_IgE_over_unstima], axis=1) 
```

```python
overlaps= pd.concat([intersects, complements], axis=1)
```

```python
overlaps.to_excel('210107 RNAseq top40 genes split.xlsx', index=False)
```
